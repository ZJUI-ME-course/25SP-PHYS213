```python
# 1
# given c1 dertaH R T1 T2 
# find c2
# 利用ln(c2/c1)=dertaH/R*(1/T1-1/T2)

from sympy import *
c1=9e-4
dertaH=2.208e-20
T1=283
T2=293
R=1.38e-23


an=dertaH/R*(1/T1-1/T2)
c2=c1*exp(an)
print(c2)
```
