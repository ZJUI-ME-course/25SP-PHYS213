```python
# 2
# 思路
# 就是先用Q相等的到Tf
# 然后用下面的公式算dertaS
# dertaS=integrate(dertaQ/T)=integrate(m*c*dertaT/T)=m*k*(Tf-Ti)
from sympy import *

kc=1e-3

m1=1
m2=5
Ta=0.8
Tb=0.3

Tf=S("Tf")

q1=m1*kc*(Ta**2-Tf**2)-m2*kc*(Tf**2-Tb**2)
Tf=(solve(q1))[1]

S=m1*kc*(Tf-Ta)+m2*kc*(Tf-Tb)

print(S)
```

```python
# 6
from sympy import *

# 能量定义修正（单位：eV → J）
E1 = -3.5 * 1.6e-19  # 基态能量
E2 = -2.8 * 1.6e-19  # 第一激发态
E3 = -1.8 * 1.6e-19  # 简并能级（简并度 g=2）

# 温度参数修正
T = 7000 + 273.15        # 开尔文温度
k = 1.38e-23             # 玻尔兹曼常数 (J/K)
beta = 1 / (k * T)       # 修正运算顺序

# 配分函数计算（含简并度）
Z = exp(-beta*E1) + exp(-beta*E2) + 3*exp(-beta*E3)

p1=1*exp(-beta*E1)/Z
p2=1*exp(-beta*E2)/Z
p3=3*exp(-beta*E3)/Z

print(p3)
U=p1*E1+p2*E2+p3*E3
print(U/1.6e-19)



print('NKln5')
```
