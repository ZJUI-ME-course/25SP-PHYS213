```python
# 1

R=8.314 #J/mol/K
Cv=3*R # 
Cp=Cv+R

p=101.3e3 #Pa
V=0.5e-3 #m**3

T=30
dertaT=10
#############################

T=T+273.15 
n=p*V/R/T
print(n)


Qx=n*Cv*dertaT
Qy=n*Cp*dertaT
print(Qx-Qy)
```

```python
# 2
ma=200  #g
mc=500
Ma=27 #g/mol Al
Mc=63.5 #g/mol Cu

n1=ma/Ma
n2=mc/Mc

Ta=76.85
Ta=Ta+273.15 # K

Tc=26.85
Tc=Tc+273.15

# 
from sympy import *

Tf=S('Tf')

q1=n1*(Tf-Ta)+n2*(Tf-Tc)
Tf=solve([q1],[Tf])[Tf]
Tf=Tf-273.15
print(Tf)

```
