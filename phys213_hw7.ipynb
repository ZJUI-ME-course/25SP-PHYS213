```python
# 2
# dertaS=n*R*ln(Vf/Vi)

# 第二问
# pi*vi=n*R*Ti
# 绝热pv**(gamma)=constant

# 第三问



from sympy import *
Vf=0.029
Vi=0.01

##############################
R=8.314 # J/(K*mol)
Ti=300
p1=100e3




n=p1*Vi/Ti/R
print(n)


dertaS1=n*R*ln(Vf/Vi)
print(dertaS1)





Ndof=3
alpha=Ndof/2
gamma=(alpha+1)/alpha



Tf=Ti*(Vi/Vf)**(gamma-1)
print(Tf)


dertaS2=n*alpha*R*ln(Tf/Ti)
print(dertaS2)


dertaS=dertaS1+dertaS2

print(dertaS)

```

```python
# 4
# Qh汤释放的热量先给热机，然后释放给环境


# 为什么低温热源吸收热量是Qc=-Tc*dertaS
# dertaSenv=integrate(dQ/T)
# 但是汤的温度是变化的，所以算出来是dertaS热=Cv*ln(Tc/Th)
# 环境温度是不变化的，所以算出来是dertaS环=Qc/Tc
# 然后极限两个S相加为0 进而得到Qc



from sympy import *

Tc=300 #k
Th=340 #K
Cv=33  #J/K

# 
Qh=Cv*(Th-Tc)
dertaS=Cv*ln(Tc/Th)


Qc=-Tc*dertaS

W=Qh-Qc 
print(W/Qh)


```
