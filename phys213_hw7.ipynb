```python
# 2
# dertaS=n*R*ln(Vf/Vi)

# 
# pi*vi=n*R*Ti
# pv**(gamma)=constant

# 



from sympy import *
Vf=0.027
Vi=0.01
R=8.314
Ti=300
p1=100e3
###############################

dertaS1=R*ln(Vf/Vi)
print(dertaS1)


Ndof=3
alpha=Ndof/2
gamma=(alpha+1)/alpha

Tf=Ti*(Vi/Vf)**(gamma-1)
print(Tf)


dertaS2=alpha*R*ln(Tf/Ti)
print(dertaS2)


dertaS=dertaS1+dertaS2

print(dertaS)

```
