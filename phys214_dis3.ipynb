```python
k=1.38e-23
T=300
from sympy import *

E0=0
E1=3/2*1e-22
E2=2*1e-22

eE0=exp(-E0/k/T)
eE1=exp(-E1/k/T)
eE2=exp(-E2/k/T)
etotal=eE0+eE1+eE2

p1=eE0/etotal
p2=eE1/etotal
p3=eE2/etotal

E0*p1+E1*p2+E2*p3

```

```python
import matplotlib.pyplot as plt

# 数据
x = [0, 1, 2, 3, 4, 5, 6]  # E_n (单位: ε)
y = [0.333, 0.25, 0.179, 0.119, 0.071, 0.036, 0.012]  # P_n

# 创建图形
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'bo', markersize=8)  # 'bo' 表示蓝色圆点

# 添加标题和标签
plt.title('1', fontsize=14)
plt.xlabel('2', fontsize=12)
plt.ylabel('3', fontsize=12)

# 设置坐标轴范围
plt.xlim(-0.5, 6.5)
plt.ylim(0, 0.5)

# 添加网格线
plt.grid(True, linestyle='--', alpha=0.7)

# 标注每个点的概率值
for i in range(len(x)):
    plt.text(x[i], y[i] + 0.01, f'{y[i]:.3f}', ha='center')

# 显示图形
plt.show()
```
