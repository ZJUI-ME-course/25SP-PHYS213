```python
na=1
nb=2
v=1
b=4e-4

from sympy import *

va=S('va')
vb=v-va

q1=na*ln(va-b*na)+nb*ln(vb)

q1_dot=diff(q1,va)
print(solve(q1_dot,va))
```
